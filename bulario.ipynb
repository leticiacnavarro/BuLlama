```python
import socket
import urllib.error
import urllib.request
from bs4 import BeautifulSoup
```

```python
def get_page(url):
    try:
        req = urllib.request.Request(url, headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/35.0.1916.47 Safari/537.36'})
        with urllib.request.urlopen(req, timeout = 5) as f:
            content = f.read().decode('utf-8')
            return BeautifulSoup(content, features="html5lib")
    except socket.timeout as e1:
        print('\tTime out')
    except urllib.error.URLError as e2:
        print('\tURL error')
    except urllib.error.HTTPError as e3:
        print('\tHTTP error')
    return None
```

```python
def get_bula_page(soup):
  bula = soup.find(id="bulaBody")
  links = bula('h2')
  for i in links:
    i.extract()
  links = bula('h3')
  for i in links:
    i.extract()
  links = bula('h4')
  for i in links:
    i.extract()
  bula_final = bula.get_text().strip().replace("_","")
  return bula_final
```

```python
def get_questions_bula(url_bula):
  bula_aux = get_page(url_bula)
  bula = bula_aux.find(id="bulaBody")
  title = bula_aux.title.get_text().split('-')[0].strip()
  elements = bula.find_all("h3")
  list_dict = []
  for el in elements:
    dict = {
        "pergunta":el.get_text().replace("?", f" {title}?"),
        "resposta":el.find_next_siblings()[0].get_text().strip().replace("\n","")
    }

    if(dict["pergunta"] != "Laborat√≥rio"):
      list_dict.append(dict)
  return list_dict
```

```python
def get_pagination(url_base, soup):
  list_pages = []
  pagination = soup.find(attrs={"class": "pagination"})
  if pagination:
    pagination = pagination("a")
    for el in pagination:
      url_pagina = url_base + el['href']
      list_pages.append(url_pagina)
  return list(dict.fromkeys(list_pages))
```

```python
def get_drug_list(url_base, soup):
  drug_list = []
  lista = soup.find(attrs={"class": "lists_list"})("a")
  for el in lista:
    url_remedio = url_base + el['href']
    drug_list.append(url_remedio)
  return list(dict.fromkeys(drug_list))
```

```python
url_base = 'https://www.bulario.com'

all_urls = []

for letter in list(map(chr, range(97, 123))):
  url_letter = f"https://www.bulario.com/alfa/{letter}/"
  all_urls.append(url_letter)
  page_letter = get_page(url_letter)
  all_urls.extend(get_pagination(url_base, page_letter))
```

```python
url_base = 'https://www.bulario.com'

all_drug_urls = []

for url in all_urls:
  page_letter = get_page(url)
  all_drug_urls.extend(get_drug_list(url_base, page_letter))
```
