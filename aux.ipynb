```python
import os
import pandas as pd

# Função para ler o arquivo de perguntas e respostas
def ler_perguntas_respostas(arquivo):
    with open(arquivo, 'r', encoding='utf-8') as file:
        linhas = file.readlines()
        perguntas_respostas = [eval(linha) for linha in linhas]
    return perguntas_respostas

# Função para ler o arquivo de texto corrido
def ler_texto_corrido(arquivo):
    with open(arquivo, 'r', encoding='utf-8') as file:
        linhas = file.readlines()
        contexto = ''.join(linhas[1:4])  # Linhas 1, 2 e 3 como contexto
        texto_corrido = ''.join(linhas)
    return contexto, texto_corrido

# Diretório contendo os arquivos
diretorio = 'anvisa'

# Lista para armazenar os prompts
prompts = []

# Iterar sobre os arquivos no diretório
for arquivo in os.listdir(diretorio):
    if arquivo.endswith("_perguntas.txt"):
        # Ler perguntas e respostas
        caminho_arquivo_perguntas = os.path.join(diretorio, arquivo)
        perguntas_respostas = ler_perguntas_respostas(caminho_arquivo_perguntas)

        # Identificar o arquivo de texto corrido correspondente
        nome_arquivo_texto = arquivo.replace("_perguntas.txt", "_texto.txt")
        caminho_arquivo_texto = os.path.join(diretorio, nome_arquivo_texto)

        # Ler contexto e texto corrido
        contexto, texto_corrido = ler_texto_corrido(caminho_arquivo_texto)

        # Criar o prompt para cada pergunta e resposta
        for par in perguntas_respostas:
            pergunta, resposta = list(par.items())[0]
            prompt = f"""[INST] <<SYS>> Você é um assistente que irá responder mensagens sobre bulas de remédio. Responda apenas o que for pedido, considerando o seguinte contexto:
{contexto.strip()} 
<</SYS>>
{pergunta}?[/INST]

{resposta}
""".strip()

            # Adicionar o prompt à lista
            prompts.append({'text': prompt})

# Criar um DataFrame do pandas
df = pd.DataFrame(prompts)

# Imprimir o DataFrame
df.iloc[0].values

```

```python
import pandas as pd

# Carregar os dataframes
bulario_df = pd.read_csv('datasets/bulario_questions.csv', delimiter=';')
bulario_plaintext_df = pd.read_csv('datasets/bulario_plain_text.csv', delimiter=';')

# Criar uma lista de prompts
prompts = []

for index, row in bulario_df.iterrows():
# Iterar sobre as colunas do dataframe bulario_df
    for column in bulario_df.columns[2:]:  # Ignorando a primeira coluna "Nome"
        print(bulario_df['Nome'][index])
        contexto = bulario_plaintext_df.loc[bulario_plaintext_df['Nome'] == bulario_df['Nome'][index], 'texto'].values[0]

        #pergunta = f'Qual é a {column} do remédio {bulario_df["Nome"][index]}?'
        pergunta = column.replace('?',f" {bulario_df['Nome'][index]}?")
        resposta_esperada = bulario_df[column][index]
        
        # Adicionar o prompt à lista de prompts
        if not pd.isna(resposta_esperada) and not pd.isna(pergunta):
            prompt = generate_prompt(pergunta, contexto.replace('\n',''), resposta_esperada)
            prompts.append(prompt)
        #print(prompt + '\n')

# Criar um novo dataframe com a coluna 'text'
prompts_df = pd.DataFrame({'text': prompts})

# Salvar o dataframe em um arquivo CSV com delimitador ;
prompts_df.to_csv('datasets/bulario_prompts.csv', sep=';', index=False)
```

```python
from bullama.bulario import web_scrapping

def get_bula_page(soup):
  bula = soup.find(id="bulaBody")
  content_except_laboratorio = []
 
  # Iterar sobre os filhos da div "bula"
  for child in bula.children:
      # Verificar se o filho é uma tag h2 com o texto "Laboratório"
      if 'Laboratório' in child.get_text():
          break  # Parar a iteração ao encontrar a tag h2 "Laboratório"
      if 'Informações Legais' in child.get_text():
          break  # Parar a iteração ao encontrar a tag h2 "Laboratório"   
      # Adicionar o conteúdo à lista
      if child.name != 'h2' and child.name != 'h3' and child.name != 'h4': 
        content_except_laboratorio.append(child)
  bula_text = ""
  for ctx in content_except_laboratorio:
     bula_text += " " + ctx.get_text().strip().replace("_","")

  return bula_text

page = web_scrapping.get_page("https://www.bulario.com/Lacril/", True)
get_bula_page(page)
```
